```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
petrol = pd.read_csv('./petrol_prices.csv')
petrol
```

```python
plt.plot(petrol['Year'],petrol['USA'])
```

```python
plt.plot(petrol['Year'],petrol['Canada'], marker='*')
```

```python
plt.plot(petrol['Year'],petrol['USA'], marker='.', label='USA')
plt.plot(petrol['Year'],petrol['Canada'], marker='.', label='Canada')
plt.legend()
```

- Can you add plot title `"Petrol prices over time in different countries"`?
- In the `fontdict`, can you set the `fontname` to `sans-serif`?
- Can you change the `fontsize` to `20` and `color` to `#d08d88`?
- Can you change the `xlabel` to `"Years"`?
- Can you change the `ylabel` to `"Price in dollars"`?
- Show the legend as well.

```python
plt.title('Petrol prices over time in different countries', fontdict = {'fontname':'sans-serif', 'fontsize':20, 'color':"#d08d88"})
plt.xlabel('Years')
plt.ylabel('Price in dollars')
plt.plot(petrol['Year'],petrol['USA'], label='USA', marker='.')
plt.plot(petrol['Year'],petrol['Canada'], label='Canada', marker='.')
plt.legend()
```

```python
plt.title('Petrol prices over time in different countries', fontdict = {'fontname':'sans-serif', 'fontsize':20, 'color':"#d08d88"})
plt.xlabel('Years')
plt.ylabel('Price in dollars')
for each in petrol:
    if each!='Year':
        plt.plot(petrol['Year'],petrol[each], label=each, marker='*')
plt.legend()
plt.plot()
```

### Simple bar graph

```python
x = ['HR', 'QA', 'Developer']
y = [40000, 50000, 90000]
plt.xlabel('Profession')
plt.ylabel('Salary')
plt.bar(x,y)
plt.show()

# Exercise: title, xlabel, ylabel
```

```python
x = ['HR', 'QA', 'Developer']
y = [40000, 50000, 90000]
bars = plt.bar(x,y)
bars[0].set_hatch('/')
bars[1].set_hatch('*')
plt.show()

# Exercise: title, xlabel, ylabel
```

### Scatter plots
